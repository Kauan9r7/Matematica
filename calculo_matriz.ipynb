Neste notebook, a matriz `A` e definida simbolicamente com `sympy`, depois seu determinante e sua matriz inversa sao calculados passo a passo.

As celulas Markdown e os comentarios abaixo foram adicionados apenas para documentar o raciocinio.


```python
# Importa as ferramentas de algebra usadas no calculo.
from sympy import symbols, Matrix, simplify, sin, cos, sqrt, trigsimp, cancel, pi

# Declara os simbolos que seram usados.
γ00 , γ11, r, m, θ, ϕ = symbols('γ00 γ11 r m θ ϕ')

# Monta a matriz A.
A = Matrix([
    [γ00, 0, 0, 0],
    [0, γ11*sin(θ)*cos(ϕ), r*cos(θ)*cos(ϕ), -r*sin(θ)*sin(ϕ)],
    [0, γ11*sin(θ)*sin(ϕ), r*cos(θ)*sin(ϕ), r*sin(θ)*cos(ϕ)],
    [0, γ11*cos(θ), -r*sin(θ), 0]
])

# Define as funcoes γ00 e γ11.
γ00 = sqrt(1-(2*m/r))
γ11 = sqrt(1/(1-(2*m/r)))

# Atribuir valores as variaveis.
"""
A = A.subs("γ00", γ00)
A = A.subs("γ11", γ11)
A = A.subs("r", 1)
A = A.subs("m", 1)
A = A.subs("θ", pi/2)
A = A.subs("ϕ", pi/2)
"""
# Exibe a matriz do problema.
display(A)
```

## Etapa 1: determinante da matriz

Aqui o determinante de `A` é calculado. O método escolhido é uma variação da Eliminação de Gauss, e `trigsimp` simplifica o resultado usando identidades trigonometricas.


```python
# Calcula o determinante de A e simplifica a expressao.
A_det = trigsimp(A.det(method="bareiss"))
# Mostra o determinante obtido.
display(A_det)
```

## Etapa 2: matriz adjunta

Depois do determinante, a proxima etapa é calcular a matriz adjunta de `A`. Ela será usada diretamente na formula da inversa. O método usado foi o método dos cofatores (expansão de Laplace).


```python
# Calcula a matriz adjunta de A e simplifica o resultado.
A_adj = trigsimp(A.adjugate(method="laplace"))
# Exibe a matriz adjunta encontrada.
display(A_adj)
```

## Etapa 3: matriz inversa

Com a adjunta e o determinante calculados, a inversa é definida pela relacao `A^{-1} = adj(A) / det(A)`. Em seguida, os elementos são simplificados para deixar a resposta final mais legivel.


```python
# Usa a relacao entre adjunta e determinante para obter a matriz inversa.
A_inv = (A_adj / A_det).applyfunc(cancel and trigsimp and simplify)
# Exibe a matriz inversa simplificada.
display(A_inv)
```

## Etapa 4: verificação do resultado

Para confirmar que a inversa foi calculada corretamente, a matriz original é multiplicada por `A_inv`. O resultado esperado é a matriz identidade.


```python
# Verifica o resultado multiplicando A pela inversa; o esperado e a identidade.
display(trigsimp(A * A_inv))
```
